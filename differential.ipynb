## 1.Hermite differential equation 
\begin{equation}
\frac{d^2y}{dx^2}-2x\frac{dy}{dx}+2ny=0
\end{equation}	
Hermite polynomial
\begin{equation}
H_n(x)=\sum_{j=0}^{N}{(-1)^j\frac{n!~(2x)^{n-2j}}{j!(n-2j)!}}
\end{equation}	

```python
def Hermite(x,n):
    if(n%2==0):# even n
        N=int(n/2)
    else:
        N=int((n-1)/2) # odd n
    Hx=sum([(-1)**j*(np.math.factorial(n)/(np.math.factorial(j)*np.math.factorial(n-2*j)))*(2*x)**(n-2*j)for j in [i for i in range(N)]]) 
    return Hx
```

```python
y=Hermite(1,1)
y
```

```python
c=['r','b','g','m','y','c'] #color
```

```python
x =np.arange(-2,2,0.1)
plt.figure(figsize = [20,6])
for n in range(2,6):
    y1=Hermite(x,n)
    plt.plot(x,y1,color=c[n],linewidth=3, label= "n= "+str(n))
    plt.xlabel("x",fontsize=18) 
plt.ylabel(r"H$_n$(x)",fontsize=18)
plt.legend()
plt.title('Hermile Polynomial',color ='r',fontsize=24)
plt.show()
```

## 2.Bessel differential equation 
\begin{equation}
x^2\frac{d^2y}{dx^2}+x\frac{dy}{dx}+(x^2-n^2)y=0
\end{equation}
Bessel polynomial
\begin{equation}
J_n(x)=\sum_{j=0}^{\infty}{(-1)^j\frac{1}{j!(n+j)!}~(x/2)^{n+2j}}
\end{equation}

```python
def Bessel(x,n):
    Jn = sum([(-1)**j*(1/(np.math.factorial(j)*np.math.factorial(n+j)))*(x/2)**(n+2*j) for j in [i for i in range(20) ]])
    return Jn 
```

```python
plt.figure(figsize = [20,6])
    
for n in range(5):
    y2=Bessel(x,n)
    plt.plot(x,y2,color=c[n],linewidth=3, label= "n= "+str(n))
plt.xlabel("x",fontsize=18) 
plt.ylabel(r"J$_n$(x)",fontsize=18)
plt.legend()
plt.title('Bessel polynomial',color='r',fontsize=24)
plt.show()
```

## 3.Laguerre differential equation 
\begin{equation}
x\frac{d^2y}{dx^2}+(1-x)\frac{dy}{dx}+ny=0
\end{equation}
Laguerre polynomial
\begin{equation}
L_n(x)=\sum_{j=0}^{n}{(-1)^j\frac{n!}{(j!)^2(n-j)!}~x^j}
\end{equation}

```python
x =np.arange(-5,10,0.1)
def Laguerre(x,n):
    Ln = sum([(-1)**j*(np.math.factorial(n)/((np.math.factorial(j))**2*np.math.factorial(n-j)))*x**j for j in [i for i in range(n) ]])
    return Ln
```

```python
plt.figure(figsize = [20,6])

for n in range(1,5):
    y3=Laguerre(x,n)
    plt.plot(x,y3,color=c[n],linewidth=3, label= "n= "+str(n))
plt.xlabel("x",fontsize=18) 
plt.ylabel(r"L$_n$(x)",fontsize=18)
plt.legend()
plt.title('Laguerre Polynomial',color='r',fontsize=24)
plt.show()
```

## 4.Legendre differential equation 
\begin{equation}
(1-x^2)\frac{d^2y}{dx^2}-2x\frac{dy}{dx}+n(n+1)y=0
\end{equation}	
Legendre polynomial
\begin{equation}
P_n(x)=\sum_{j=0}^{N}{(-1)^j\frac{(2n-2j)!~x^{n-2j}}{2^j~j!(n-2j)(n-j)!}}
\end{equation}	

```python
x =np.arange(-1,1,0.1)
def Legendre(x,n):
    if(n%2==0):# even n
        N=int(n/2)
    else:
        N=int((n-1)/2) # odd n
    Pn = sum([(-1)**j*(np.math.factorial(2*n-2*j)/(2**j*np.math.factorial(j)*np.math.factorial(n-2*j)*np.math.factorial(n-j)))*x**(n-2*j) for j in [i for i in range(N) ]])
    return Pn
```

```python
plt.figure(figsize = [20,6])

for n in range(2,6):
    y4=Legendre(x,n)
    plt.plot(x,y4,color=c[n],linewidth=3, label= "n= "+str(n))
plt.xlabel("x",fontsize=18) 
plt.ylabel(r"P$_n$(x)",fontsize=18)
plt.legend()
plt.title('Legendre polynomial',color='r',fontsize=24)
plt.show()
```
